MAPS




```
pip install folium
```

```
import folium as fo
```

```
map=fo.Map()
```

```
map
```

```
x=fo.FeatureGroup(name="My Map")
```

```
x.add_child(fo.Marker(location=[17.852369,79.377783],popup = 'My Home',icon=fo.Icon(color='darkred')))
map.add_child(x)
```

```
y=fo.FeatureGroup(name="My Map")
y.add_child(fo.Marker(location=[17.9926,79.5544],popup = 'Home2',icon=fo.Icon(color='darkblue')))
map.add_child(y)
```

```
z=fo.FeatureGroup(name="My Map")
z.add_child(fo.Marker(location=[17.851845,79.380144],popup = 'Home',icon=fo.Icon(color='darkgreen')))
map.add_child(z)
```
